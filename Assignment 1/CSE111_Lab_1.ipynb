```python
#1
word=input()
upper=0
lower=0
for i in word:
    if i.isupper():
        upper+=1
    else:
        lower+=1
if upper>lower:
    new=word.upper()
    print(new)
else:
    new=word.lower()
    print(new)
```

```python
#5
password=input()
lw=False
up=False
digit=False
sp=False
res=''
for i in password:
    if i.isupper():
        up=True
    elif i.islower():
        lw=True
    elif i.isdigit():
        digit=True
    elif i in '_ , $ , #, @':
        sp=True
        
if lw==True and up==True and digit==True and sp==True:
    print('OK')
if lw==False:
    res=res+'Lowercase character missing,'
if up==False:
    res=res+'Uppercase character missing,'
if digit==False:
    res=res+'Digit missing,'
if sp==False:
    res=res+'Special character missing,'
print(res[:-1])

```

# LIST

```python
#1
count=1
list1=[]
list2=[]
while count>0:
    num=input()
    if num=="STOP":
        break 
    else:
        list1.append(int(num))
        count+=1
for i in list1:
    if i not in list2:
        list2.append(i)
for i in list2:
    freq=list1.count(i)
    print(f"{i} - {freq} times")
        
        
```

```python
#3
list1=[]
count=1
list2=[]

while count>0:
    num=input()
    if num=='STOP':
        break
    else:
        list1.append(int(num))
        count+=1
length=len(list1)-1

for i in range(len(list1)):
    if i==len(list1)-1:
        break
    else:
        diff=abs(list1[i]-list1[i+1])
        list2.append(diff)
        
if max(list2)>length:
    print("Not UB Jumper")
else:
    print("UB Jumper")
        
    
    
```

```python
#1
dict2={}
dict3={}
newdict={}
for i in range(2):
    if i==0:
        dict0=input().split(',')
        for j in dict0:
            new=j.split(':')
            dict2[new[0]]=int(new[1])
    else:
        dict0=input().split(',')
        for j in dict0:
            new=j.split(':')
            dict3[new[0]]=int(new[1])
print(dict2)
print(dict3)
for key,value in dict2.items():
    if key in dict3:
        newdict[key]=dict2[key]+dict3[key]
    else:
        newdict[key]=value
for key,value in dict3.items():
    if key not in dict2:
        newdict[key]=value
print(newdict)

new=[]       
for values in newdict.values():
    if values not in new:
        new.append(values)
new.sort()
print('Values: ',tuple(new))
```

# Function

```python
#1
def BMI(height,weight):
    nheight=height/100
    bmi=weight/nheight**2

    if bmi<18.5:
        print(f"Score is {round(bmi,1)}. You are Underweight")
    elif 18.5<bmi<24.9:
        print(f"Score is {round(bmi,1)}. You are Normal")
    elif 25<bmi<30:
        print(f"Score is {round(bmi,1)}. You are Overweight")
    else:
        print(f"Score is {round(bmi,1)}. You are Obese")
BMI(int(input()),int(input()))
```

```python
#2
def divisor(low,high,div):
    add=0
    for i in range(low,high):
        if i%div==0:
            add+=i
    return add
print(divisor(int(input()),int(input()),int(input())))
```

```python
#task 5
def palindrome(string):
    new=''
    for i in string:
        if i!=' ':
            new+=i
    if new==new[::-1]:
        print('Palindrome')
    else:
        print('Not a palindrome')
        
palindrome(input())
```

```python
#task 6
def capitalize(string):
    new_one=''
    for i in range(len(string)):
        if i==0:
            new_one+=string[i].upper()
        else:
            new_one+=string[i]
        #unless i==0-2
        if string[i-2]=='.' or string[i-2]=='!' or string[i-2]=='?':
            new_one+=string[i].upper()
        else:
             new_one+=string[i]
        if string[i]=='i' and string[i-1]==' ' and string[i+1]==' ':
            new_one+=string[i]
            
        
    
    
    
    
    

    
print(capitalize(input()))
```
